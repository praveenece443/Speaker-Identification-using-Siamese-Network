```python
import os
import shutil
```

```python
path = "dataset/"

for ids in os.listdir(path):
    subfolders = path + ids
    for folders in os.listdir(subfolders):
        subs = subfolders + '/' + folders
        for filepath in os.listdir(subs):
            files = subs + '/' + filepath
            names = files.split('/')[-1]
            shutil.move(files, subfolders + '/' + names)
```

```python
path = "dataset/"

for ids in os.listdir(path):
    subfolders = path + ids
    for folders in os.listdir(subfolders):
        subs = subfolders + '/' + folders
        isDirectory = os.path.isdir(subs)
        if isDirectory:
            os.rmdir(subs)
```

```python
path = "dataset/"
i = 0
for ids in os.listdir(path):
    subfolders = path + ids
    print(subfolders)
    shutil.move(subfolders, path + '/' + str(i))
    i += 1
```
