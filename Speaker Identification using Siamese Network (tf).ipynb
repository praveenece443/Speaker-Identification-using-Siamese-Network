```python
def audio2vector(file_path, max_pad_len=1600):    
    X, sample_rate = librosa.load(file_path, res_type='kaiser_fast') 
    mfccs = librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40)   
    pad_width = max_pad_len - mfccs.shape[1]
    print(pad_width)
    mfcc = np.pad(mfccs, pad_width=((0, 0), (0, pad_width)), mode='constant')
    return mfcc
```

```python
%%time

path = "dataset/"
data = []
labels = []
for folders in os.listdir(path):
    files = path + folders
    for file in os.listdir(files):
        try:
            audios = files + '/' + file
            audio_data = audio2vector(audios)
            print(audio_data.shape)
            data.append(audio_data)
            labels.append(folders)
        except:
            continue
```

```python
data = np.array(data)
labels = np.array(labels)
print(data.shape,labels.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
labels = le.fit_transform(labels)
```

```python
def make_pairs(data, labels):
    pairData = []
    pairLabels = []
    numClasses = len(np.unique(labels))
    idx = [np.where(labels == i)[0] for i in range(0, numClasses)]
    for idxA in range(len(data)):
        currentAudio = data[idxA]
        label = labels[idxA]
        idxB = np.random.choice(idx[label])
        posData = data[idxB]
        pairData.append([currentAudio, posData])
        pairLabels.append([1])
        negIdx = np.where(labels != label)[0]
        negData = data[np.random.choice(negIdx)]
        pairData.append([currentAudio, negData])
        pairLabels.append([0])
    return (np.array(pairData), np.array(pairLabels))
```

```python
# prepare the positive and negative pairs
print("[INFO] preparing positive and negative pairs...")
(pairdata, pairlabels) = make_pairs(data, labels)
```

```python
trainx, testx, trainy, testy = train_test_split(pairdata, pairlabels, test_size=0.25, random_state=42)
print(trainx.shape, testx.shape, trainy.shape, testy.shape)
```

```python
trainy = trainy.astype(np.float32)
testy = testy.astype(np.float32)
```

```python
from tensorflow.keras import backend as K
from tensorflow.keras import layers
from tensorflow.keras.layers import Activation,Subtract 
from tensorflow.keras.layers import Input, Lambda, Dense, Dropout, Flatten, Conv1D, MaxPooling1D
from tensorflow.keras.models import Model, Sequential
import tensorflow as tf

config = tf.compat.v1.ConfigProto()
config.gpu_options.allow_growth=True
sess = tf.compat.v1.Session(config=config)
```

```python
def build_base_network(input_shape):
    encoder = Sequential()
    encoder.add(layers.Conv1D(64, 32, padding='same', activation='relu', input_shape=input_shape))
    encoder.add(layers.BatchNormalization())
    encoder.add(layers.SpatialDropout1D(0.5))
    encoder.add(layers.MaxPool1D(4, 4))

    # Further convs
    encoder.add(layers.Conv1D(2*32, 3, padding='same', activation='relu'))
    encoder.add(layers.BatchNormalization())
    encoder.add(layers.SpatialDropout1D(0.5))
    encoder.add(layers.MaxPool1D())

    encoder.add(layers.Conv1D(4 * 16, 3, padding='same', activation='relu'))
    encoder.add(layers.BatchNormalization())
    encoder.add(layers.SpatialDropout1D(0.5))
    encoder.add(layers.MaxPool1D())

    encoder.add(layers.Conv1D(3 * 8, 3, padding='same', activation='relu'))
    encoder.add(layers.BatchNormalization())
    encoder.add(layers.SpatialDropout1D(0.5))
    
    encoder.add(layers.Conv1D(1 * 4, 3, padding='same', activation='relu'))
    encoder.add(layers.BatchNormalization())
    encoder.add(layers.SpatialDropout1D(0.5))

    encoder.add(layers.GlobalMaxPool1D())

    encoder.add(layers.Dense(16000))
    return encoder
```

```python
input_dim = trainx.shape[2:]

audio_a = Input(shape=input_dim)
audio_b = Input(shape=input_dim)
```

```python
base_network = build_base_network(input_dim)

feat_vecs_a = base_network(audio_a)
feat_vecs_b = base_network(audio_b)
```

```python
#distance = Lambda(euclidean_distance)([feat_vecs_a, feat_vecs_b])
distance = Subtract()([feat_vecs_a, feat_vecs_b])
distance = Lambda(lambda x: K.abs(x))(distance)
```

```python
outputs = Dense(1, activation="sigmoid")(distance)
model = Model(inputs=[audio_a, audio_b], outputs=outputs)
```

```python
model.compile(loss="binary_crossentropy", optimizer="rmsprop",metrics=["accuracy"])
# train the model
print("[INFO] training model...")
history = model.fit([trainx[:, 0], trainx[:, 1]], trainy[:],
                    validation_data=([testx[:, 0], testx[:, 1]], testy[:]),
                    batch_size=256, 
                    epochs=1000)
```

```python
model.save("voice_siamese.h5")
```

```python
model.evaluate([testx[:, 0], testx[:, 1]],testy[:])
```
